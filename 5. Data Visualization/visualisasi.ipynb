```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

```python
data_origin = dataset["origin"].value_counts().to_dict()

data_pie = data_origin.values()
data_label = data_origin.keys()

colors = sns.color_palette("pastel")
```

```python
plt.pie(
    x=data_pie,
    labels=data_label,
    colors=colors,
    autopct="%.0f%%"
)
plt.title("Origin")
plt.show()
```

```python
sns.histplot(
    data=dataset,
    x="mpg",
    bins=30,
    kde=True
)
plt.show()
```

```python
sns.boxplot(
    data=dataset,
    x="mpg",
)
plt.show()
```

```python
sns.barplot(
    data=dataset,
    x="origin",
    y="mpg",
    hue="origin"
    # hue="cylinders"
)
plt.show()
```

```python
sns.lineplot(
    data=dataset,
    x="model_year",
    y="mpg",
    hue="origin"
)

plt.show()
```

```python
sns.scatterplot(
    data=dataset,
    x="displacement",
    y="mpg",
    hue="origin"
)

plt.show()
```
