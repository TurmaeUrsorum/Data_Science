## üî• Distribusi yang sering dipakai Data Scientist

### 1. **Normal (Gaussian) Distribution**

* Paling populer ‚Üí banyak fenomena nyata **mendekati normal**.
* Dipakai untuk:

  * Tinggi/berat badan
  * Nilai ujian
  * Noise pada data (error dalam regresi)
* Kenapa penting? ‚Üí banyak metode statistik & ML (misalnya regresi linear, ANOVA) mengasumsikan error berdistribusi normal.

---

### 2. **Binomial & Bernoulli**

* Dipakai untuk data **biner / kategori 2 kelas**.
* Contoh:

  * Click vs tidak click (CTR di ads)
  * Menang vs kalah
  * Sukses vs gagal
* Binomial = jumlah sukses dalam beberapa percobaan Bernoulli.

---

### 3. **Poisson Distribution**

* Untuk data **jumlah kejadian** dalam periode waktu tertentu.
* Contoh:

  * Jumlah customer datang per jam
  * Jumlah error server dalam 1 hari
  * Jumlah panggilan masuk call center

---

### 4. **Exponential Distribution**

* Berhubungan sama Poisson ‚Üí waktu tunggu antar kejadian.
* Contoh:

  * Waktu antara kedatangan pelanggan
  * Umur pakai mesin sampai rusak

---

### 5. **Uniform Distribution**

* Semua nilai dalam rentang punya peluang sama.
* Contoh:

  * Sampling random
  * Random initialization di algoritma ML

---

### 6. **Log-Normal, Gamma, Beta, t-Distribution**

* Banyak dipakai di data real-life:

  * **Log-Normal**: harga saham (karena nggak bisa negatif)
  * **Gamma/Exponential**: modeling waktu hidup (survival analysis)
  * **Beta**: modeling probabilitas (karena terikat di 0‚Äì1)
  * **t-Distribution**: dipakai saat sample kecil

---

## üéØ Singkatnya

* Kalau data **kontinu real** ‚Üí sering pakai **Normal**
* Kalau data **hitungan kejadian** ‚Üí **Poisson / Binomial**
* Kalau data **waktu antar-event** ‚Üí **Exponential**
* Kalau data **probabilitas (0‚Äì1)** ‚Üí **Beta**

## Distribusi Bernoulli

```python
from scipy.stats import bernoulli

koin = bernoulli.rvs(size=1000, p=0.5)

sns.histplot(koin)
plt.show()
```

```python
koin
```

## Distribusi Binomial

```python
from scipy.stats import binom

koin = binom.rvs(size=1000, n=10, p=0.5)

sns.histplot(koin)
plt.show()
```

```python
koin
```

## Distribusi Poisson

```python
from scipy.stats import poisson

cacat = poisson.rvs(size = 1000, mu = 2)

#Membuat plot
sns.histplot(cacat)
plt.show()
```

```python
cacat
```

# Distribusi Kontinu

## Distribusi Seragam

```python
from scipy.stats import uniform

uniform_data = uniform.rvs(size = 1000, loc = 0, scale = 9)

#Membuat plot
sns.histplot(uniform_data)
plt.show()
```

## Distirbusi Normal

```python
from scipy.stats import norm

data_nilai = norm.rvs(size = 1000, loc = 70, scale = 10)

#Membuat plot
sns.histplot(data_nilai)
plt.show()
```

## Distirbusi Eksponensial

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Buat dataset palsu (tinggi badan dalam cm)
np.random.seed(42)
data = np.random.normal(loc=170, scale=10, size=500)  # Normal(170, 10^2)

# 2. Histogram data
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label="Data")

# 3. Fit distribusi normal ke data
mu, std = stats.norm.fit(data)  
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)  # pdf normal

# 4. Plot distribusi teoritis
plt.plot(x, p, 'r', linewidth=2, label=f"Normal fit: Œº={mu:.2f}, œÉ={std:.2f}")
plt.legend()
plt.show()
```
