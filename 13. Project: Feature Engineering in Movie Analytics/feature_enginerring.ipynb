## Data Preparation

```python
movie_data["release_date"] = pd.to_datetime(movie_data["release_date"])
```

#### cek nilai data

### Filter Data

```python
# movie_data[movie_data["status"] == "Released"]
# movie_data[movie_data["original_language"] == "en"]

english_release = movie_data[(movie_data["status"] == "Released") & (movie_data["original_language"] == "en")]
```

```python
english_release.shape
```

### Sorting Data

```python
english_release.sort_values("budget", ascending=False).head(10)
```

```python
english_release.sort_values(["budget", "revenue"], ascending=[False, False]).tail(10)
```

### Duplication Data dan Missing Data

```python
english_release.duplicated().sum()
```

```python
english_release[english_release.duplicated(subset="original_title", keep=False)].sort_values("original_title")
```

```python
english_release.drop_duplicates(inplace=True)
```

```python
english_release.duplicated().sum()
```

```python
english_release.info()
```

```python
english_release.isna().sum()
```

```python
english_release["tagline"].fillna("No Tagline", inplace=True)
english_release["belongs_to_collection"].fillna("No Data", inplace=True)
english_release["homepage"].fillna("No Data", inplace=True)
english_release["poster_path"].fillna("No Data", inplace=True)
```
