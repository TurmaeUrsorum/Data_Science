## Feature Engineering

#### Agregation

```python
rating = pd.read_csv("./ratings_small.csv")
rating.head()
```

```python
rating.shape
```

```python
rating["timestamp"] = pd.to_datetime(rating["timestamp"], unit="s")
```

```python
rating.sample(10)
```

```python
rating.info()
```

```python
rating.duplicated().sum()
```

```python
rating.isna().sum()
```

```python
rating.groupby("movieId")["rating"].agg(["mean", "count", "median"])
```

```python
rating.groupby("movieId").agg({
    "userId": pd.Series.nunique,
    "rating": ["mean", "count", "median"]
})
```

```python
movies_rating = rating.groupby("movieId")["rating"].agg(["mean", "median", "count"])
```

```python
movies_rating.head()
```

#### Merging Data or Joining Data

```python
english_release.head()
```

```python
english_release_rating = english_release.merge(
    movies_rating,
    left_on="id",
    right_on="movieId",
    how="inner"
)
```

```python
english_release_rating
```

#### Feature Engineering

```python
english_release_rating.revenue
```

```python
english_release_rating.budget
```

```python
english_release_rating["profit"] = english_release_rating["revenue"] - english_release_rating["budget"]
```

```python
english_release_rating[["original_title", "profit", "runtime"]].sort_values("profit", ascending=False).head(10)
```

```python
english_release_rating["long_duration_movie"] = (english_release_rating["runtime"] >= 60 * 3).astype("int")
```

```python
english_release_rating[english_release_rating["long_duration_movie"] == 1].head()
```

```python
english_release_rating["title_length"] = english_release_rating["original_title"].str.len()
```

```python
english_release_rating[english_release_rating["title_length"] <= 10].head()
```

```python
english_release_rating["duration_hours"] = english_release_rating["runtime"] / 60
```

```python
english_release_rating
```

```python
english_release_rating["year_released"] = english_release_rating["release_date"].dt.year
```

```python
english_release_rating["favorite_movie"] = (english_release_rating["median"] >= 3).astype("int")
```

```python
english_release_rating["favorite_movie"].value_counts()
```

```python
year_release = english_release_rating.groupby("year_released").size()
```

```python
year_release.sort_values(ascending=False).head(10)
```
